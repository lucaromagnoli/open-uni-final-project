```python
import os
import pandas as pd
from pandas.io.json import json_normalize
import ast
```

```python
infile = 'verniani_2020-02-08-23-06.csv'
```

```python
scraped_dir = '/Users/luca.romagnoli/Dropbox/RossiRei/scraped-data/'
processed_dir = '/Users/luca.romagnoli/Dropbox/RossiRei/processed-data/'
inpath = os.path.join(scraped_dir, infile)
outpath = os.path.join(processed_dir, infile.replace('csv', 'xlsx'))
```

```python
df = pd.read_csv(inpath, encoding='utf-16', sep='\t')
```

```python
df['image_url'] = (
    df['images'].apply(lambda x: ast.literal_eval(x))
    .apply(lambda x: f"http://rossi-rei-data.s3.us-east-2.amazonaws.com/manufacturers/pictures/{x[0]['path']}"))
```

```python
df.to_excel(outpath, index=False)
```
